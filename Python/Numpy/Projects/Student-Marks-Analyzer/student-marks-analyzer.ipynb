```python
import numpy as np
```

```python
marks = np.array([[85, 78, 92],
                  [88, 76, 79],
                  [90, 91, 85],
                  [70, 65, 72],
                  [60, 58, 64]
                ])
```

```python
print("Student Marks:\n", marks)
```

```python
# Student wise average
student_avg = np.mean(marks, axis=1)
print("\nAverage marks per student:\n", student_avg)
```

```python
# subject wise average
subject_avg = np.mean(marks, axis=0)
print("\nAverage marks per student:\n", subject_avg)
```

```python
# Highest and Lowest marks
print("Highest marks:", np.max(marks))
print("Lowest marks:", np.min(marks))
```

```python
# Topper?
total_marks = np.sum(marks, axis=1)
topper_index = np.argmax(total_marks)

print("Topper student index: ", topper_index)
print("Topper total marks: ", total_marks[topper_index])
```

```python
# Pass/Fail
pass_status = np.all(marks >= 40, axis=1)
print("Pass status of students: ", pass_status)
```

```python
passed_count = np.sum(pass_status)
failed_count = len(pass_status) - passed_count

print("Passed students: ", passed_count)
print("Failed Students: ", failed_count)
```
