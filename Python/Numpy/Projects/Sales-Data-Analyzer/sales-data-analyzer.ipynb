```python
import numpy as np
```

```python
sales = np.array([
    [12000, 15000, 13000, 16000, 17000, 18000, 17500, 16500, 15500, 14500, 13500, 12500],
    [10000, 11000, 10500, 11500, 12000, 12500, 13000, 12800, 12200, 11800, 11200, 10800],
    [8000, 8500, 9000, 9500, 10000, 10500, 11000, 10800, 10200, 9800, 9200, 8700]
])
```

```python
total_sales = np.sum(sales, axis=1)
print("Total sales per product: ", total_sales)
```

```python
average_sales = np.mean(sales, axis=1)
print("Average sales per product:", average_sales)
```

```python
best_product = np.argmax(total_sales)
print("Best performing product index:", best_product)
```

```python
worst_product = np.argmin(total_sales)
print("Worst performing product index:", worst_product)
```

```python
monthly_sales = np.sum(sales, axis=0)
print("Monthly total sales:", monthly_sales)
```

```python
best_month = np.argmax(monthly_sales)
print("Best sales month index:", best_month)
```

```python
high_sales = sales > 15000
print(high_sales)
```
